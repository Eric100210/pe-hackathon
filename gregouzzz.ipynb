```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
df = pd.DataFrame(pd.read_excel('data2022.xls'))
mask = df['year']==2022
df1 = df.copy()
df1=df1[mask]
mask2=df1['Healthy life expectancy at birth'].notna()
df1=df1[mask2]
df1.head()
X=np.array(df1['Healthy life expectancy at birth'])
Y=np.array(df1['Life Ladder'])
lm = linear_model.LinearRegression()
lm.fit(X.reshape(-1,1), Y.reshape(-1,1))

print("Le coeff dir est:", lm.coef_)
print("La constante est:", lm.intercept_)

plt.scatter(X,Y)
def f(h):
    return lm.coef_*h + lm.intercept_

H=np.linspace(0,75,100)
F=[int(f(h)) for h in H]
plt.plot(H,F)
```
